```python
import random 

#define parameters
min_val = 1
max_val = 100000000

nums = [random.randint(min_val, max_val) for _ in range(20)]
```

```python
def conquer(left, right):
    temp = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] > right[j]:
            temp.append(left[i])
            i += 1
        else:
            temp.append(right[j])
            j += 1

    temp.extend(left[i:])
    temp.extend(right[j:])

    return temp

def mergeSort(array):
    if len(array) <= 1:
        return array
    mid = len(array) // 2
    left = mergeSort(array[:mid])
    right = mergeSort(array[mid:])
    
    return conquer(left, right)

mergeSort(nums)
```
